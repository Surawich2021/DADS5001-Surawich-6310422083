```python
# Load the revenue dataset
uri_rv2016 = 'DADS_5001/Revenue2016.xlsx'
df_rv2016 = pd.read_excel(uri_rv2016)  

uri_rv2017 = 'DADS_5001/Revenue2017.xlsx'
df_rv2017 = pd.read_excel(uri_rv2017)

uri_rv2018 = 'DADS_5001/Revenue2018.xlsx'
df_rv2018 = pd.read_excel(uri_rv2018)  

uri_rv2019 = 'DADS_5001/Revenue2019.xlsx'
df_rv2019 = pd.read_excel(uri_rv2019)

uri_rv2020 = 'DADS_5001/Revenue2020.xlsx'
df_rv2020 = pd.read_excel(uri_rv2020)  

uri_rv2021 = 'DADS_5001/Revenue2021.xlsx'
df_rv2021 = pd.read_excel(uri_rv2021)

uri_rv2022 = 'DADS_5001/Revenue2022.xlsx'
df_rv2022 = pd.read_excel(uri_rv2022)
```

```python
df_rv2016.isnull().sum()
df_rv2016.Name = df_rv2016.Name.astype('category')
df_rv2016.info()

df_rv2017.isnull().sum()
df_rv2017.Name = df_rv2017.Name.astype('category')
df_rv2017.info()

df_rv2018.isnull().sum()
df_rv2018.Name = df_rv2018.Name.astype('category')
df_rv2018.info()

df_rv2019.isnull().sum()
df_rv2019.Name = df_rv2019.Name.astype('category')
df_rv2019.info()

df_rv2020.isnull().sum()
df_rv2020.Name = df_rv2020.Name.astype('category')
df_rv2020.info()

df_rv2021.isnull().sum()
df_rv2021.Name = df_rv2021.Name.astype('category')
df_rv2021.info()

df_rv2022.isnull().sum()
df_rv2022.Name = df_rv2022.Name.astype('category')
df_rv2022.info()
```

```python
df_rv2016 = df_rv2016.loc[ : , ['Name','Total 2016'] ]   # Choose columns
df_rv2016

df_rv2017 = df_rv2017.loc[ : , ['Name','Total 2017'] ]   # Choose columns
df_rv2017

df_rv2018 = df_rv2018.loc[ : , ['Name','Total 2018'] ]   # Choose columns
df_rv2018

df_rv2019 = df_rv2019.loc[ : , ['Name','Total 2019'] ]   # Choose columns
df_rv2019

df_rv2020 = df_rv2020.loc[ : , ['Name','Total 2020'] ]   # Choose columns
df_rv2020

df_rv2021 = df_rv2021.loc[ : , ['Name','Total 2021'] ]   # Choose columns
df_rv2021

df_rv2022 = df_rv2022.loc[ : , ['Name','Total 2022'] ]   # Choose columns
df_rv2022
```

```python
df_outer = pd.merge( df_rv2016, df_rv2017, left_on='Name', right_on='Name', how='outer' )
df_outer = pd.merge( df_outer, df_rv2018, left_on='Name', right_on='Name', how='outer' )
df_outer = pd.merge( df_outer, df_rv2019, left_on='Name', right_on='Name', how='outer' )
df_outer = pd.merge( df_outer, df_rv2020, left_on='Name', right_on='Name', how='outer' )
df_outer = pd.merge( df_outer, df_rv2021, left_on='Name', right_on='Name', how='outer' )
df_outer = pd.merge( df_outer, df_rv2022, left_on='Name', right_on='Name', how='outer' )
df_outer
```

```python
df_outer.isnull().sum()
```

```python
colm = ['Total 2016','Total 2017','Total 2018','Total 2019','Total 2020','Total 2021','Total 2022']
df_outer[colm] = df_outer[colm].fillna(0)
df_outer.isnull().sum()
```

```python
df_outer['Total'] = df_outer['Total 2016'] + df_outer['Total 2017'] + df_outer['Total 2018'] + df_outer['Total 2019'] + df_outer['Total 2020'] + df_outer['Total 2021'] + df_outer['Total 2022']
df_outer.head()
```

```python
df_outer = df_outer.loc[ : , ['Name','Total'] ]   # Choose columns
df_outer
```

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
df_outer.sort_values('Total', ascending=False, inplace=True)
df_outer
```

```python
top10 = df_outer.head(10)
top10
```

```python
top10 = top10.reset_index(drop=True)
top10
```

```python
x = np.arange(1,11)
xdata = []
ydata = []
for i in range(0,10):
    xdata.append(top10.iloc[(i,0)])
    ydata.append(top10.iloc[(i,1)])

plt.figure(figsize=[20,10])
ax = plt.axes(xticks=x) # กำหนดให้ขีดวางตรงทุกค่าตำแน่งที่ป้อนเข้าไป (คือ 1 ถึง 10)
ax.set_xticklabels(xdata,rotation=10,fontname='Tahoma',fontsize='13') # ใส่ชื่อจังหวัดลงไปแทนตัวเลข 1 ถึง 10
ax.set_title(u'TOP10 อุทยานแห่งชาติรายได้สูงสุด',fontname='Tahoma',fontsize='13')
ax.set_xlabel(u'อุทยาน',fontname='Tahoma',fontsize='13')
ax.set_ylabel(u'รายได้สะสมปี2016-2022 (พันล้านบาท)',labelpad=-4,fontname='Tahoma',fontsize='13')

plt.savefig('TOP10 Revenue.png', dpi=1000, transparent=True)
plt.bar(x,ydata,color='b')
plt.show()
```
